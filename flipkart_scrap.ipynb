```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python
url='https://www.flipkart.com/'

# driver=webdriver.Chrome()
driver=webdriver.Chrome()
driver.maximize_window()
time.sleep(2)
driver.get(url)
time.sleep(2)

driver.find_element(By.CLASS_NAME,'b3wTlE').click()
time.sleep(2)

search_bar=driver.find_element(By.NAME,'q')
search_bar.send_keys('sg cricket bats')
search_bar.send_keys(Keys.ENTER)
time.sleep(3)

names, prices,review, no_of_reviews=[],[],[],[]
for i in range(5):
    time.sleep(3)
    containers=driver.find_elements(By.CLASS_NAME,'RGLWAk')
    for container in containers:
        names.append(container.find_element(By.CLASS_NAME,'pIpigb').text)
        try:
            prices.append(container.find_element(By.CLASS_NAME,'hZ3P6w').text)
        except:
            prices.append('Not able to scrap')
        try:
            review.append(container.find_element(By.CLASS_NAME,'MKiFS6').text)
        except:
            review.append("No review available")
        try:
            no_of_reviews.append(container.find_element(By.CLASS_NAME,'PvbNMB').text)
        except:
            no_of_reviews.append("Not available")
            

    try:
        driver.find_elements(By.CLASS_NAME,'jgg0SZ')[-1].click()
    except:
        print("not able to proceed")

    


driver.quit()

n={'bat names':names, 'price':prices,'review':review, 'review_counts':no_of_reviews}
df=pd.DataFrame(n)
print(df)
df.to_excel('Bats.xlsx',index=False)
```
